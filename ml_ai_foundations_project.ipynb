## Data Wrangling and Tidying

In this section, I clean the dataset and prepare it for exploratory data analysis.

The focus is on checking the shape of the data, reviewing data types, selecting useful columns, and handling missing values in the selected dataset.

```python
print("Original dataset shape:")
print(hpi.shape)
```

```python
hpi.dtypes
```

```python
selected_columns = [
    "Date",
    "RegionName",
    "AreaCode",
    "AveragePrice",
    "SalesVolume"
]

hpi_clean = hpi[selected_columns].copy()

hpi_clean.head()
```

I selected a smaller group of columns for the first stage of analysis. These columns include the date, location information, the average house price, and sales volume.

This creates a simpler working dataset while still keeping the main variables needed for exploratory analysis and future machine learning preparation.

```python
hpi_clean.isnull().sum()
```

```python
hpi_clean = hpi_clean.dropna()

print("Cleaned dataset shape:")
print(hpi_clean.shape)
```

Rows with missing values in the selected columns were removed. This keeps the working dataset simple and ensures the remaining data is complete for exploratory analysis.

```python
hpi_clean.dtypes
```

```python
hpi_clean.head()
```

```python
hpi_clean.info()
```

```python
hpi_clean["Year"] = hpi_clean["Date"].dt.year
hpi_clean["Month"] = hpi_clean["Date"].dt.month

hpi_clean.head()
```

I created separate `Year` and `Month` columns from the `Date` column. These new columns will make the date information easier to use during analysis and future machine learning preparation.

### Wrangling Summary

The dataset was cleaned by selecting the most relevant columns, checking missing values, removing incomplete rows from the selected data, and confirming the data types.

The cleaned dataset, `hpi_clean`, is now ready for exploratory data analysis.

## Exploratory Data Analysis

In this section, I explore the cleaned dataset using summary statistics and visualizations.

The goal is to understand the distribution of average house prices, how prices change over time, and whether there are visible relationships between selected variables.

```python
hpi_clean.describe()
```

The summary statistics show the range and spread of the numeric columns in the cleaned dataset. The main variable of interest is `AveragePrice`, which will later be used as the target variable for machine learning preparation.

```python
average_price_mean = hpi_clean["AveragePrice"].mean()
average_price_median = hpi_clean["AveragePrice"].median()
average_price_min = hpi_clean["AveragePrice"].min()
average_price_max = hpi_clean["AveragePrice"].max()
average_price_range = average_price_max - average_price_min

print(f"Mean average price: £{average_price_mean:,.2f}")
print(f"Median average price: £{average_price_median:,.2f}")
print(f"Minimum average price: £{average_price_min:,.2f}")
print(f"Maximum average price: £{average_price_max:,.2f}")
print(f"Range: £{average_price_range:,.2f}")
```

The mean and median help show the typical average house price in the dataset. Comparing the mean and median can also show whether very high or very low prices are affecting the overall average.

```python
hpi_clean[["AveragePrice", "SalesVolume", "Year", "Month"]].corr()
```

The correlation table shows the relationships between numeric variables. This helps identify whether variables such as `SalesVolume`, `Year`, or `Month` have a relationship with `AveragePrice`.

```python
import matplotlib.pyplot as plt

plt.hist(hpi_clean["AveragePrice"], bins=30)
plt.title("Distribution of Average House Prices")
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Average House Prices

The histogram shows that average house prices are not evenly distributed. Most observations are concentrated at the lower to middle price range, while a smaller number of observations have much higher average prices.

This creates a right-skewed distribution, meaning there are some high-value areas that pull the range upward. This suggests that location may be an important factor when preparing the dataset for machine learning.


```python
average_price_by_year = hpi_clean.groupby("Year")["AveragePrice"].mean()

average_price_by_year.plot()
plt.title("Average House Price Over Time")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

### Average House Price Over Time

The line chart shows a clear upward trend in average house prices over time. Prices generally increased from the mid-1990s to 2026, although there are periods where the trend slows or dips.

One noticeable dip appears around the late 2000s, which may reflect a wider housing market slowdown. Overall, the chart suggests that time-based features such as `Year` and `Month` could be useful when preparing the data for future machine learning modeling.


```python
top_10_areas = hpi_clean.groupby("RegionName")["AveragePrice"].mean().sort_values(ascending=False).head(10)

top_10_areas.plot(kind="bar")
plt.title("Top 10 Areas by Average House Price")
plt.xlabel("Area")
plt.ylabel("Average Price")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Top 10 Areas by Average House Price

The bar chart shows the areas with the highest average house prices in the dataset. The most expensive areas are mainly London-based, including Kensington and Chelsea, Westminster, Camden, Hammersmith and Fulham, and the City of London.

This suggests that location is likely to be one of the most important features for predicting average house prices. Area-based variables such as RegionName or AreaCode should therefore be considered when preparing the dataset for machine learning.

```python
plt.scatter(hpi_clean["SalesVolume"], hpi_clean["AveragePrice"], alpha=0.5)
plt.title("Sales Volume vs Average House Price")
plt.xlabel("Sales Volume")
plt.ylabel("Average Price")
plt.show()
```

### Sales Volume vs Average House Price

The scatter plot compares `SalesVolume` with `AveragePrice`. The chart does not show a simple linear relationship between the two variables.

Many high average price observations appear where sales volume is relatively low, while areas with very high sales volumes tend to have more moderate average prices. This suggests that `SalesVolume` alone may not be a strong predictor of average house price, but it may still provide useful information when combined with other features such as location and time.


### EDA Summary

The exploratory analysis shows that average house prices vary widely across the dataset. The distribution of prices is right-skewed, with a smaller number of very high-value areas.

House prices also show a strong upward trend over time, suggesting that date-based features may be useful for future modeling. Location also appears important, as the highest average prices are concentrated in specific areas, especially London boroughs.

Overall, the EDA suggests that `Year`, `Month`, `RegionName`, `AreaCode`, and `SalesVolume` may be useful features to consider when preparing the dataset for machine learning.
