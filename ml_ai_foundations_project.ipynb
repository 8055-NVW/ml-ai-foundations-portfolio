## Loading and Checking the Dataset

In this section, I load the UK House Price Index dataset and inspect its structure to confirm that it is suitable for machine learning preparation.

The key checks are:

- whether the CSV file loads correctly
- how many rows and columns the dataset contains
- what each row appears to represent
- whether the dataset contains the variables needed for analysis
- whether the target variable, `AveragePrice`, is available

```python
import pandas as pd

hpi = pd.read_csv("data/UK-HPI-full-file-2026-04.csv")

hpi["Date"] = pd.to_datetime(hpi["Date"], dayfirst=True)

hpi.head()
```

```python
print("Dataset shape:")
print(hpi.shape)
```

```python
hpi.info()
```

```python
hpi.describe()
```

```python
print("Earliest date:", hpi["Date"].min())
print("Latest date:", hpi["Date"].max())

hpi[["Date", "RegionName", "AreaCode"]].head(10)
```

```python
area_month_duplicates = hpi.duplicated(subset=["Date", "AreaCode"]).sum()

print("Duplicate Date + AreaCode combinations:", area_month_duplicates)
```

### Data Suitability Check

The dataset loaded successfully and contains the main variables needed for this project, including `Date`, `RegionName`, `AreaCode`, `AveragePrice`, and `SalesVolume`.

Each row appears to represent house price information for one UK area during one month. The combination of `Date` and `AreaCode` is unique, which suggests the dataset has an appropriate level of detail for area-based house price analysis.

The target variable for the future machine learning task will be `AveragePrice`, making this a supervised regression problem.

## Data Wrangling and Tidying

In this section, I clean the dataset and prepare it for exploratory data analysis.

The focus is on checking the shape of the data, reviewing data types, selecting useful columns, and handling missing values in the selected dataset.

```python
print("Original dataset shape:")
print(hpi.shape)
```

```python
hpi.dtypes
```

```python
selected_columns = [
    "Date",
    "RegionName",
    "AreaCode",
    "AveragePrice",
    "SalesVolume"
]

hpi_clean = hpi[selected_columns].copy()

hpi_clean.head()
```

I selected a smaller group of columns for the first stage of analysis. These columns include the date, location information, the average house price, and sales volume.

This creates a simpler working dataset while still keeping the main variables needed for exploratory analysis and future machine learning preparation.

```python
hpi_clean.isnull().sum()
```

```python
hpi_clean = hpi_clean.dropna()

print("Cleaned dataset shape:")
print(hpi_clean.shape)
```

Rows with missing values in the selected columns were removed. This keeps the working dataset simple and ensures the remaining data is complete for exploratory analysis.

```python
hpi_clean.dtypes
```

```python
hpi_clean.head()
```

```python
hpi_clean.info()
```

```python
hpi_clean["Year"] = hpi_clean["Date"].dt.year
hpi_clean["Month"] = hpi_clean["Date"].dt.month

hpi_clean.head()
```

I created separate `Year` and `Month` columns from the `Date` column. These new columns will make the date information easier to use during analysis and future machine learning preparation.

### Wrangling Summary

The dataset was cleaned by selecting the most relevant columns, checking missing values, removing incomplete rows from the selected data, and confirming the data types.

The cleaned dataset, `hpi_clean`, is now ready for exploratory data analysis.
